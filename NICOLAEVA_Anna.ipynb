```python
import pandas as pd
import numpy as np
import seaborn as sns
from scipy.stats import chi2_contingency
import matplotlib.pyplot as plt

# Create a sample 2x2 contingency table
data = {
    'Deces': [14, 22],
    'Externat': [367, 579]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Perform the chi-squared test
chi2, p, _, _ = chi2_contingency(df)

# Create a heatmap to visualize the contingency table
plt.figure(figsize=(6, 4))

# Specify custom y-axis labels
y_labels = ['M', 'F']

# Create the heatmap with custom y-axis labels
sns.heatmap(df, annot=True, cmap="YlGnBu", fmt="d", cbar=False, yticklabels=y_labels)

# Add the chi-squared test result to the plot
plt.text(0.75, 0.9, f"Chi-Squared Statistic: {chi2:.2f}\nP-value: {p:.4f}",
         fontsize=12, ha="center", va="center", transform=plt.gca().transAxes)

# Set plot labels and title
plt.xlabel('Result')
plt.ylabel('Sex')
plt.title('2x2 Contingency Table Heatmap')

plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import mannwhitneyu


# Create a box plot to compare the two groups
plt.figure(figsize=(6, 4))
sns.boxplot(x='Rezultat', y='Varsta', data=df_final)
plt.xlabel('Rezultat')
plt.ylabel('Varsta')
plt.title('Comparative Evaluation Between Two Groups')

# Perform Mann-Whitney U test
group_A = df_final[df_final['Rezultat'] == 'externat']['Varsta']
group_B = df_final[df_final['Rezultat'] == 'decedat']['Varsta']
statistic, p_value = mannwhitneyu(group_A, group_B, alternative='two-sided')

# Add the Mann-Whitney U test result to the plot
result_text = f'Mann-Whitney U Test\np-value = {p_value:.4f}'
plt.text(0.5, 0.85, result_text, fontsize=10, ha='center', va='center', transform=plt.gca().transAxes)

plt.show()
```
